```python
import json
import re
from pathlib import Path
from typing import Optional

from tqdm import tqdm

from utils import get_llm, get_response_text


INDEX_HEADER_RE = re.compile(r'^##\s+Содержание\s*$', re.MULTILINE)
INDEX_LINE_RE = re.compile(r'^\s*(\d+\.|[a-z]\.)\s+\[([^\]]+)\]\(#([^\)]+)\)\s*$')
HEADING_RE = re.compile(r'^(#{3,4})\s+(.+)$', re.MULTILINE)


def extract_index(article: str) -> tuple[str, list[str]]:
    match = INDEX_HEADER_RE.search(article)
    if match is None:
        raise ValueError('Не найден блок "## Содержание"')

    lines = article[match.start():].splitlines()
    index_lines, parts = [], []
    for line in lines:
        stripped = line.rstrip()
        if not stripped:
            if parts:
                break
            index_lines.append(stripped)
            continue
        if stripped.startswith('## ') and stripped != '## Содержание' and index_lines:
            break
        index_lines.append(stripped)
        parsed = INDEX_LINE_RE.match(stripped)
        if parsed:
            marker, title, _ = parsed.groups()
            parts.append(f'{marker} {title}')

    if not parts:
        raise ValueError('Не удалось распарсить пункты содержания')

    return '\n'.join(index_lines), parts


def part_has_children(parts: list[str], index: int) -> bool:
    current = parts[index].split('.')[0]
    if not current.isdigit():
        return False

    for next_part in parts[index + 1:]:
        next_marker = next_part.split('.')[0]
        if next_marker.isdigit():
            return False
        return True
    return False


def validate_part_article(part_article: str, part: str) -> None:
    headings = HEADING_RE.findall(part_article.strip())
    if len(headings) != 1:
        raise ValueError('В ответе должен быть ровно один заголовок')

    marker = part.split('.')[0]
    expected_level = '###' if marker.isdigit() else '####'
    heading_level, _ = headings[0]
    if heading_level != expected_level:
        raise ValueError(f'Ожидался заголовок {expected_level}, получен {heading_level}')

    body = HEADING_RE.sub('', part_article, count=1).strip()
    if not body:
        raise ValueError('После заголовка нет текста')
    if re.search(r'^(?:[-*]|\d+\.|[a-z]\.)\s+', body, re.MULTILINE):
        raise ValueError('В тексте раздела не должно быть списков')
    if re.search(r'^#{1,6}\s+', body, re.MULTILINE):
        raise ValueError('В тексте раздела появился лишний заголовок')


def generate_with_retries(llm, prompt: str, validator, retries: int = 4) -> str:
    last_error = None
    for _ in range(retries):
        response = llm.invoke(prompt)
        text = get_response_text(response)
        try:
            validator(text)
            return text
        except ValueError as error:
            last_error = error
    raise ValueError(f'Не удалось получить корректный ответ: {last_error}')


def build_article(concept: str, llm=None) -> tuple[str, list[str]]:
    llm = llm or globals().get('gemini')
    if llm is None:
        raise ValueError('Сначала выполните ячейку выбора рабочей модели')

    index_prompt = index_prompt_template.format(concept=concept)
    intro_article = generate_with_retries(
        llm,
        index_prompt,
        lambda text: extract_index(text),
    )
    index_block, parts = extract_index(intro_article)

    sections = [intro_article.strip()]
    for part in tqdm(parts, desc=concept, unit='part'):
        part_prompt = part_prompt_template.format(
            concept=concept,
            part=part,
            index=index_block,
        )
        part_article = generate_with_retries(
            llm,
            part_prompt,
            lambda text, current_part=part: validate_part_article(text, current_part),
        )
        sections.append(part_article.strip())

    return '\n\n'.join(sections), parts


def generate_articles(
    output_dir: Path,
    concept_ids: Optional[list[str]] = None,
    skip_existing: bool = True,
    stop_on_error: bool = False,
) -> tuple[list[Path], list[tuple[str, str, str]]]:
    with open(Path.cwd().parent / 'ontology.json', 'r', encoding='utf-8') as file:
        ontology = json.load(file)

    concepts = [
        (concept_id, concept)
        for concept_id, concept in ontology['labels'].items()
        if concept_id.startswith('Q') and (concept_ids is None or concept_id in concept_ids)
    ]

    output_dir.mkdir(parents=True, exist_ok=True)
    saved_paths = []
    failed = []
    for concept_id, concept in tqdm(concepts, desc='articles', unit='article'):
        output_path = output_dir / f'{concept_id}.md'
        if skip_existing and output_path.exists():
            print(f'{concept_id}: {concept} -> пропуск, файл уже существует')
            saved_paths.append(output_path)
            continue

        try:
            article, parts = build_article(concept, llm=globals().get('gemini'))
        except Exception as error:
            print(f'{concept_id}: {concept} -> ошибка: {error}')
            failed.append((concept_id, concept, str(error)))
            if stop_on_error:
                raise
            continue
        with open(output_path, 'w', encoding='utf-8') as file:
            file.write(article)
        print(f'{concept_id}: {concept} -> {len(parts)} разделов')
        saved_paths.append(output_path)

    return saved_paths, failed


# Тестовый прогон на одном понятии. Для полной генерации уберите concept_ids или передайте нужный список.
saved_paths, failed = generate_articles(
    output_dir=Path.cwd().parent / 'advanced_papers',
    skip_existing=True,
)
print(f'Успешно: {len(saved_paths)}, ошибок: {len(failed)}')
failed[:10]
```
